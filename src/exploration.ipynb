```python

from ghcnd.utils import *
```

```python
ghcnd = GHCND()

stations = pd.read_parquet(ghcnd.datalake_root / 'clean' / 'stations.parquet')
weather_dataset = ds.dataset(ghcnd.datalake_root / 'curated' / 'combined')

# Print the data schema: each line defines a column
print(weather_dataset.schema)
```

```python
us_stations = stations[stations['station_id'].str[:2] == 'US']
ca_stations = stations[stations['station_id'].str[:2] == 'CA']
mx_stations = stations[stations['station_id'].str[:2] == 'MX']
stations = pd.concat([us_stations, ca_stations, mx_stations])
stations.sample(10)
```

```python
filter = (pc.field('year') >= 2000) & (pc.field('year') <= 2025)

weather = weather_dataset.filter(filter).to_table().to_pandas()
```
